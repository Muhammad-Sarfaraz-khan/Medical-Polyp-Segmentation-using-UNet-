```python
def generate_polyp_mask(h=256, w=256):
    mask = np.zeros((h, w), dtype=np.uint8)

    for _ in range(np.random.randint(1, 3)):
        center = (np.random.randint(50, w-50), np.random.randint(50, h-50))
        axes = (np.random.randint(20, 60), np.random.randint(20, 60))
        angle = np.random.randint(0, 180)
        cv2.ellipse(mask, center, axes, angle, 0, 360, 255, -1)

    return mask

def generate_colonoscopy_image(mask):

    img = np.random.normal(loc=120, scale=20, size=(256, 256, 3)).astype(np.uint8)

    colored_polyp = cv2.applyColorMap(mask, cv2.COLORMAP_JET)
    polyp_region = (colored_polyp * (mask[..., None] > 0)).astype(np.uint8)

    return cv2.addWeighted(img, 0.8, polyp_region, 0.2, 0)


print("Generating dataset...")

for i in range(N):
    mask = generate_polyp_mask()
    img  = generate_colonoscopy_image(mask)

    cv2.imwrite(f"data/images/img_{i:03}.png", cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    cv2.imwrite(f"data/masks/mask_{i:03}.png", mask)

print("Dataset created successfully!")
print("Folders:")
print("- data/images (synthetic colonoscopy images)")
print("- data/masks  (binary segmentation masks)")
```
